```python
import ast
import pandas as pd

# Select relevant columns
col4 = df.iloc[:, 4]  # assuming this is the 'link' column
col5 = df.iloc[:, 5]  # assuming some metadata or filter column

# Filter out rows with 'Recipes1M' in col5
mask = ~col5.str.contains("Recipes1M", na=False)
filtered_df = df[mask].copy()

# Extract base domain from link
def extract_base_link(link):
    if pd.isna(link):
        return None
    return link.split('/')[0]

# Apply extraction on filtered DataFrame
filtered_df['base_link'] = filtered_df['link'].apply(extract_base_link)

# Filter where base link is www.cookbooks.com
cookbooks_df = filtered_df[filtered_df['base_link'] == 'www.cookbooks.com']

# Extract and parse the ingredient lists
ingrs2 = [ast.literal_eval(i) for i in cookbooks_df['ingredients'].dropna().tolist()]
directions2 = [i for i in cookbooks_df['directions'].dropna().tolist()] 
original = [i for i in cookbooks_df['ingredients'].dropna().tolist()]

print(f"{len(ingrs2)}")

print(f"{len(directions2)}")

print(type(directions2[0]))
```

```python
import re

def clean_ingredient(ingredient):
    # Remove fractions and numbers (e.g., "1", "1/2", "2.5")
    ingredient = re.sub(r'\b\d+([\/\.]\d+)?\b', '', ingredient)

    # Common measurement units to remove
    units = [
        "teaspoons?", "tsp", "tablespoons?", "tbsp", "cups?", "ounces?", "oz",
        "pounds?", "lb", "grams?", "g", "kilograms?", "kg", "milliliters?", "ml",
        "liters?", "l", "pinch", "clove", "cloves", "slices?", "dash", "cans?", 
        "packages?", "bunch", "stalks?", "heads?", "pieces?", "sticks?", "inches?"
    ]
    units_pattern = r'\b(?:' + '|'.join(units) + r')\b'
    ingredient = re.sub(units_pattern, '', ingredient, flags=re.IGNORECASE)

    # Remove extra spaces
    ingredient = re.sub(r'\s+', ' ', ingredient).strip()

    return ingredient
```

```python
def clean_ingredients(data):
    return [[clean_ingredient(item) for item in sublist] for sublist in data]


cleaned_ingrs = clean_ingredients(ingrs2)

cleaned_ingrs_ = random.sample(cleaned_ingrs, 200)
print(cleaned_ingrs_)
```

```python
flattened_list = [str(item) for item in cleaned_ingrs]
len(flattened_list)
```

```python
# Create a DataFrame with directions2 and flattened_list
data = {'Directions': directions2, 'Ingredients': flattened_list, 'Ingredients_original': original}
df_combined = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df_combined.to_csv("directions_and_ingredients.csv", index=False)
```

```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
import torch

print(torch.cuda.is_available())  # True if GPU is available
print(torch.cuda.get_device_name(0))  # Get GPU name if available


# Encode labels
unique_labels = list(set(flattened_list))
label2id = {label: idx for idx, label in enumerate(unique_labels)}
id2label = {idx: label for label, idx in label2id.items()}
df_combined['label'] = df_combined['Ingredients'].map(label2id)

# Load tokenizer and tokenize input
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize_function(example):
    return tokenizer(example["Ingredients_original"], padding="max_length", truncation=True)

# Convert to Hugging Face Dataset
dataset = Dataset.from_pandas(df_combined[['Ingredients_original', 'label']])
dataset = dataset.map(tokenize_function, batched=True)
dataset = dataset.rename_column("label", "labels")
dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'labels'])

# Split dataset
dataset = dataset.train_test_split(test_size=0.2)
train_dataset = dataset['train']
eval_dataset = dataset['test']
```

```python
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=len(unique_labels))

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    num_train_epochs=3,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer
)

trainer.train()
```

```python
def predict(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    predicted_class = torch.argmax(outputs.logits, dim=1).item()
    return id2label[predicted_class]

# Example
print(predict("2 cups of chopped onions"))
```
