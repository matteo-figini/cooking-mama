```python
# Returns 1 if vegan, 0 otherwise
def check_ingredients(ingredient_string):
    ingredients_list = ast.literal_eval(ingredient_string)
    
    for row in ingredients_list:
        # Remove punctuation and numbers, make it all lowercase
        row = re.sub(r'[^\w\s]', '', row).lower()  # Remove punctuation and make it lowercase
        row = re.sub(r'\d+', '', row)       # Remove numbers
        row = row.split(" ")
        
        for ingredient in row:
            if ingredient in non_vegan_ingredients:
                return 0
    return 1
```

```python
 df["vegan"] = df["ingredients"].apply(check_ingredients)
```

```python
df.to_csv("vegan.csv", index=False)
```
